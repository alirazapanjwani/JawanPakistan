```python
#7. Mark sheet with grades
English = float(input("Enter marks of English : "))
Maths = float(input("Enter marks of Maths : "))
Urdu = float(input("Enter marks of Urdu : "))
Science = float(input("Enter marks of Science : "))
Computer = float(input("Enter marks of Computer : "))

Total = English + Maths + Urdu + Science + Computer
percentage = (Total / 500) * 100

print("Total Marks:", Total)
print("Percentage:", percentage)

if percentage >= 80:
    print("Grade: A+")
elif percentage >= 70:
    print("Grade: A")
elif percentage >= 60:
    print("Grade: B")
elif percentage >= 50:
    print("Grade: C")
else:
    print("Grade: Fail")
```
