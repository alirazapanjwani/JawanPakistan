```python
#1. Calculator (Addition, Subtraction, Multiplication, Division, Power)
# Simple Calculator

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print("Choose operation:")
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")
print("5. Power")

choice = input("Enter choice (1-5): ")

if choice == '1':
    print("Result:", num1 + num2)
elif choice == '2':
    print("Result:", num1 - num2)
elif choice == '3':
    print("Result:", num1 * num2)
elif choice == '4':
    if num2 != 0:
        print("Result:", num1 / num2)
    else:
        print("Cannot divide by zero")
elif choice == '5':
    print("Result:", num1 ** num2)
else:
    print("Invalid choice")
     
```

```python
#2. Check if there is any numeric value in a list using for loop
my_list = ["apple", "banana", 10, "cat"]

found = False

for item in my_list:
    if isinstance(item, (int, float)):
        found = True
        break

if found:
    print("Numeric value exists in the list.")
else:
    print("No numeric value found.")
     
```

```python
#3. Add a key to a dictionary

my_dict = {"name": "Ali", "age": 20}

# Adding new key
my_dict["city"] = "Karachi"

print("Updated Dictionary:", my_dict)
```

```python
#4. Sum All Numeric Items in a Dictionary
my_dict = {"a": 10, "b": 20, "c": "Hello", "d": 5}

total = 0

for value in my_dict.values():
    if isinstance(value, (int, float)):
        total += value

print("Sum of numeric values:", total)
```

```python
#5. Identify Duplicate Values from List
my_list = [1, 2, 3, 2, 4, 5, 1, 6]

duplicates = []

for item in my_list:
    if my_list.count(item) > 1 and item not in duplicates:
        duplicates.append(item)

print("Duplicate values:", duplicates)
```

```python
#6. Check if a Given Key Exists in a Dictionary
my_dict = {"name": "Ali", "age": 20, "city": "Karachi"}

key = input("Enter key to check: ")

if key in my_dict:
    print("Key exists in dictionary.")
else:
    print("Key does not exist in dictionary.")
```
